### Preparing the data ( a neural network expects to process contiguous batches of data)

```python
## Encoding the integer sequences via multi-hot encoding
import numpy as np
def vectorize_sequences(sequences, dimension=10000):
    results = np.zeros((len(sequences), dimension)) #Creates an all-zero matrix of shape (len(sequences), dimension)
    for i, sequence in enumerate(sequences):
        for j in sequence:
            results[i, j] = 1.#Sets specific indices of results[i] to 1s
    return results
x_train = vectorize_sequences(train_data)
x_test = vectorize_sequences(test_data) 
```

```python
y_train = np.asarray(train_labels).astype("float32")
y_test = np.asarray(test_labels).astype("float32")
```

### Building model

```python
from tensorflow import keras

from tensorflow.keras import layers #provides various building blocks for constructing neural networks,
                                    #such as Dense, Conv2D, LSTM, etc.
# densely connected (Dense) layers
model = keras.Sequential([          #layers are stacked one after another in a linear sequence.
layers.Dense(16, activation="relu"), # 16 units, RelU activation
layers.Dense(16, activation="relu"),
layers.Dense(1, activation="sigmoid")
])

##output = relu(dot(input, W) + b)
```

### Compiling the model

```python
model.compile(optimizer="rmsprop", #RMSProp (Root Mean Square Prop)
loss="binary_crossentropy",        #suitable for binary classification problems.
metrics=["accuracy"])
```

### Validating:  Setting aside a validation set

```python
x_val = x_train[:10000]
partial_x_train = x_train[10000:]
y_val = y_train[:10000]
partial_y_train = y_train[10000:]
```

### Training  model

```python
history = model.fit(partial_x_train, #This object has a member history, which is a dictionary 
                                     #containing data about everything that happened during training
partial_y_train,
epochs=20,
batch_size=512,
validation_data=(x_val, y_val))
```

```python
history_dict = history.history
```

### Plotting the training and validation loss

```python
import matplotlib.pyplot as plt
history_dict = history.history
loss_values = history_dict["loss"]
val_loss_values = history_dict["val_loss"]
epochs = range(1, len(loss_values) + 1)
plt.plot(epochs, loss_values, "bo", label="Training loss")     # "bo" is for "blue dot."
plt.plot(epochs, val_loss_values, "b", label="Validation loss")# "b" is for "solid blue line."

plt.title("Training and validation loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

### Plotting the training and validation accuracy

```python
plt.clf() # Clear the figure
acc = history_dict["accuracy"]
val_acc = history_dict["val_accuracy"]
plt.plot(epochs, acc, "bo", label="Training acc")
plt.plot(epochs, val_acc, "b", label="Validation acc")
plt.title("Training and validation accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

# LAB 6

```python
from tensorflow.keras.datasets import imdb
(train_data, train_labels), _ = imdb.load_data(num_words=10000)

train_data = vectorize_sequences(train_data)
model = keras.Sequential([
layers.Dense(16, activation="relu"),
layers.Dense(16, activation="relu"),
layers.Dense(1, activation="sigmoid")
])
model.compile(optimizer="rmsprop",
loss="binary_crossentropy",
metrics=["accuracy"])
history_original = model.fit(train_data, train_labels,
epochs=20, batch_size=512, validation_split=0.4)
```

```python
import matplotlib.pyplot as plt
history_dict = history_original.history
loss_values = history_dict["loss"]
val_loss_values = history_dict["val_loss"]
epochs = range(1, len(loss_values) + 1)
plt.plot(epochs, loss_values, "bo", label="Training loss")     # "bo" is for "blue dot."
plt.plot(epochs, val_loss_values, "b", label="Validation loss")# "b" is for "solid blue line."

plt.title("Training and validation loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

### Lower-capacity model version

```python
model = keras.Sequential([
layers.Dense(4, activation="relu"),
layers.Dense(4, activation="relu"),
layers.Dense(1, activation="sigmoid")
])
model.compile(optimizer="rmsprop",
loss="binary_crossentropy",
metrics=["accuracy"])
history_smaller_model = model.fit(
train_data, train_labels,
epochs=20, batch_size=512, validation_split=0.4)
```

```python
import matplotlib.pyplot as plt
history_dict = history_smaller_model.history
loss_values = history_dict["loss"]
val_loss_values = history_dict["val_loss"]
epochs = range(1, len(loss_values) + 1)
plt.plot(epochs, loss_values, "bo", label="Training loss")     # "bo" is for "blue dot."
plt.plot(epochs, val_loss_values, "b", label="Validation loss")# "b" is for "solid blue line."

plt.title("Training and validation loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plotting validation loss for history_original
plt.plot(history_original.history['val_loss'], label='Original Model')
# Plotting validation loss for history_smaller_model
plt.plot(history_smaller_model.history['val_loss'], label='Smaller Model')

# Adding labels and title
plt.xlabel('Epochs')
plt.ylabel('Validation Loss')
plt.title('Validation Loss Comparison')
plt.legend()

# Display the plot
plt.show()
```

### Smaller model starts overfitting after six epochs, while the original starts overfitting after four!

### Version of the model with higher capacity

```python
model = keras.Sequential([
    layers.Dense(512, activation="relu"),
    layers.Dense(512, activation="relu"),
    layers.Dense(1, activation="sigmoid")
])
model.compile(optimizer="rmsprop",
    loss="binary_crossentropy",
    metrics=["accuracy"])
history_larger_model = model.fit(
    train_data, train_labels,
    epochs=20, batch_size=512, validation_split=0.4)
```

```python
import matplotlib.pyplot as plt

# Plotting validation loss for history_original
plt.plot(history_original.history['val_loss'], label='Original Model')
# Plotting validation loss for history_smaller_model
plt.plot(history_larger_model.history['val_loss'], label='Larger Model')

# Adding labels and title
plt.xlabel('Epochs')
plt.ylabel('Validation Loss')
plt.title('Validation Loss Comparison')
plt.legend()

# Display the plot
plt.show()
```

### The bigger model starts overfitting almost immediately, after just one epoch

 ### Adding L2 weight regularization to the model

```python
from tensorflow.keras import regularizers
model = keras.Sequential([
    layers.Dense(16,
        kernel_regularizer=regularizers.l2(0.002),
        activation="relu"),
    layers.Dense(16,
        kernel_regularizer=regularizers.l2(0.002),
        activation="relu"),
    layers.Dense(1, activation="sigmoid")
])
model.compile(optimizer="rmsprop",
    loss="binary_crossentropy",
    metrics=["accuracy"])
history_l2_reg = model.fit(
train_data, train_labels,
epochs=20, batch_size=512, validation_split=0.4)
```

```python
import matplotlib.pyplot as plt

# Plotting validation loss for history_original
plt.plot(history_original.history['val_loss'], label='Original Model')
# Plotting validation loss for history_smaller_model
plt.plot(history_l2_reg.history['val_loss'], label='L2-regularized Model')

# Adding labels and title
plt.xlabel('Epochs')
plt.ylabel('Validation Loss')
plt.title('Validation Loss Comparison')
plt.legend()

# Display the plot
plt.show()
```

### the model with L2 regularization has become much more resistant to overfitting than the reference model,
### even though both models have the same number of parameters.

## ADDING DROPOUT

```python
model = keras.Sequential([
    layers.Dense(16, activation="relu"),
    layers.Dropout(0.5),
    layers.Dense(16, activation="relu"),
    layers.Dropout(0.5),
    layers.Dense(1, activation="sigmoid")
])
model.compile(optimizer="rmsprop",
        loss="binary_crossentropy",
        metrics=["accuracy"])
history_dropout = model.fit(
        train_data, train_labels,
        epochs=20, batch_size=512, validation_split=0.4)
```

```python
import matplotlib.pyplot as plt

# Plotting validation loss for history_original
plt.plot(history_original.history['val_loss'], label='Original Model')
# Plotting validation loss for history_smaller_model
plt.plot(history_dropout.history['val_loss'], label='Dropout Model')

# Adding labels and title
plt.xlabel('Epochs')
plt.ylabel('Validation Loss')
plt.title('Validation Loss Comparison')
plt.legend()

# Display the plot
plt.show()
```
